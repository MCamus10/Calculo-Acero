### 1. Perfil H soldado


```python
import modulo_secciones_acero as seccion
import numpy as np
```

```python
#Perfil H simétrico/asimétrico
h = 1044 #mm
bfs = 600 #mm
bfi = 600 #mm
tfs = 28 #mm
tfi = 28 #mm
tw = 12 #mm

perfilh2 = seccion.Perfilhsoldado(h, bfs, bfi, tfs, tfi, tw)
print(f"A: {perfilh2.A()} mm2")
print(f"Ix: {perfilh2.Ix()/10**6} mm4")
print(f"Sx sup: {perfilh2.Sx_sup()/10**3} mm3")
print(f"Sx inf: {perfilh2.Sx_inf()/10**3} mm3")
print(f"rx: {perfilh2.rx()} mm")
print(f"Zx: {perfilh2.Zx()/10**3} mm3")
print(f"Iy: {perfilh2.Iy()/10**6} mm4")
# print(f"Sy: {perfilh2 .Sy()/10**3} mm3")
print(f"ry: {perfilh2.ry()} mm")
print(f"Zy: {perfilh2.Zy()/10**3} mm3")
print(f"ia: {perfilh2.ia()} mm")
print(f"it: {perfilh2.it()} mm")
print(f"X1: {perfilh2.X1()} MPa")
print(f"X2: {perfilh2.X2()*10**8} (1/MPa)^2")
print(f"J: {perfilh2.J()/10**4} mm4")
print(f"Cw: {perfilh2.Cw()/10**12} mm6")
```
