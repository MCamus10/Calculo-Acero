## Propiedades de secciones t√≠picas de acero de acuerdo con manual ICHA

### 1. Perfil H soldado


```python
import moduloseccionesacero as seccion
import numpy as np
```

```python
h = 1044 #mm
bf = 600 #mm
tf = 28 #mm
tw = 12 #mm

perfilh = seccion.Perfilhsoldado(h, bf, tf, tw)
print(f"A: {perfilh.A()} mm2")
print(f"Ix: {perfilh.Ix()/10**6} mm4")
print(f"Sx: {perfilh.Sx()/10**3} mm3")
print(f"rx: {perfilh.rx()} mm")
print(f"Zx: {perfilh.Zx()/10**3} mm3")
print(f"Iy: {perfilh.Iy()/10**6} mm4")
print(f"Sy: {perfilh.Sy()/10**3} mm3")
print(f"ry: {perfilh.ry()} mm")
print(f"Zy: {perfilh.Zy()/10**3} mm3")
print(f"ia: {perfilh.ia()} mm")
print(f"it: {perfilh.it()} mm")
print(f"X1: {perfilh.X1()} MPa")
print(f"X2: {perfilh.X2()*10**8} (1/MPa)^2")
print(f"J: {perfilh.J()/10**4} mm4")
print(f"Cw: {perfilh.Cw()/10**12} mm6")
```
